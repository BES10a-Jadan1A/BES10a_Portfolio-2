```python
print("== |CHEMICAL REACTOR SAFETY CHECKER| ==")

try:
    # User input
    temperature = float(input("Enter reactor temperature (°C): "))
    pressure = float(input("Enter reactor pressure (atm): "))

    print("\n-•- ANALYZING SYSTEM STATUS -•-")

    # Temperature condition
    if temperature < 67:
        print("Temperature Status: Reactor temperature is too low.")
        print("Reaction rate may become inefficient.")

    elif temperature >= 68 and temperature <= 100:
        print("Temperature Status: Reactor temperature is NORMAL.")

        # Nested condition for pressure
        if pressure >= 1 and pressure <= 5:
            print("Pressure Status: SAFE operating pressure.")
            print("System Condition: Reactor operating normally.")

        elif pressure > 5:
            print("Pressure Status: HIGH pressure detected.")
            print("Warning: Monitor reactor carefully.")

        else:
            print("Pressure Status: Pressure too low.")
            print("Check for possible leaks.")

    else:
        print("Temperature Status: DANGEROUS temperature!")
        print("Emergency cooling system required.")

        if pressure > 5:
            print("CRITICAL ALERT: High temperature and pressure detected!")
            print("Possible reactor explosion risk!")

except:
    print("ERROR: Please enter numeric values only.")
```
